```python
###Hangman Game
name = input("Please enter you name :")
print(" Hello " +  name + " Welcome to the game. I will ask you to find the secret word. let's start if you are ready...")

word = "python"
guess = ""

chance = 10

while chance > 0:
    letter = 0

    for character in word:
        if character in guess:
            print(character)
        else:
            print("_")
            letter += 1

    if letter == 0:
        print(" Yess! You won :) ")
        break

    user = input("Please enter letter :")
    guess += user

    if user not in word:
        chance -= 1
        print("incorret !")
        print(f"your remaining chance {chance} ")

        if chance == 0:
            print("unfortunately... your luck is over ")
```
